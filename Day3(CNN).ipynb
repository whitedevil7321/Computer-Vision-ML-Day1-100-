```python
import numpy as np 
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical

import random
```

```python
np.random.seed(0)
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
assert(x_train.shape[0]==y_train.shape[0]),"The number of images is not equal to the number of labels"
assert(x_test.shape[0]==y_test.shape[0]),"The number of images is not equal to the number of labels"
assert(x_train.shape[1:]==(28,28)),"The Shape of the image is not 28*28"
assert(x_test.shape[1:]==(28,28)),"The Shape of the image is not 28*28"
```

```python
num_of_samples=[]
cols=5
num_classes=10

fig,axes=plt.subplots(nrows=num_classes,ncols=cols,figsize=(10,10))
fig.tight_layout()
for i in range(cols):
    for j in range(num_classes):
        x_selected=x_train[y_train==j]
        axes[j][i].imshow(x_selected[random.randint(0,len(x_selected-1)),:,:],cmap=plt.get_cmap("gray"))
        if i==2:
            axes[j][i].set_title(str(j))
            num_of_samples.append(len(x_selected))
        
        
```

```python
print(num_of_samples)
```

```python

plt.figure(figsize=(12,4))
plt.bar(range(len(num_of_samples)),num_of_samples)
```
